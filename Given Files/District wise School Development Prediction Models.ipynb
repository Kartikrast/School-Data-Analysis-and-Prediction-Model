```python
X = df[['school_type','state_of_your_school','shifts','total_number_of_students','infrastructure_of_your_school',
        'availability_of_water_at_school','water_and_handwashing_facilities_with_soap_near_toilets','capacity','yil_2023',
        'yil_2022','yil_2021','yil_2020','yil_2019','yil_2018','yil_2017','yil_2016','yil_2015','yil_2014',
        'yil_2013','yil_2012','yil_2011','yil_2010']]
y = df['district']
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# Scaling the numerical features
numeric_features = []
for col in X.columns:
    if(X[col].dtype == "float" or X[col].dtype == "int64"):
        numeric_features.append(col)
X[numeric_features] = scaler.fit_transform(X[numeric_features])
```

```python
from sklearn.model_selection import train_test_split
Xtrain , Xtest , ytrain , ytest = train_test_split(X,y,test_size=0.20,random_state=42) 
```

```python
from sklearn.linear_model import LogisticRegression
# instantiate the model (using the default parameters)
logreg = LogisticRegression(random_state=16)
# fit the model with data
logreg.fit(Xtrain, ytrain)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
log_pred = logreg.predict(Xtest)
```

```python
from sklearn.metrics import classification_report
log_acc = accuracy_score(ytest, log_pred)
log_cl = classification_report(ytest, log_pred)
print(f"Accuracy Score of Logistic Regression is : {log_acc}")
print(f"Classification Report : \n{log_cl}")
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix
gnb = GaussianNB()
gnb.fit(Xtrain, ytrain)
```

```python
gnb_pred = gnb.predict(Xtest)
```

```python
gnb_acc = accuracy_score(ytest, gnb_pred)
clf_gnb = classification_report(ytest, gnb_pred)
print(f"Accuracy Score of Naive Bayes is : {gnb_acc}")
print(f"Classification Report : \n{clf_gnb}")
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rd = RandomForestClassifier()
rd.fit(Xtrain, ytrain)
```

```python
rd_pred = rd.predict(Xtest)
```

```python
clf_rd = classification_report(ytest, rd_pred)
rd_acc = accuracy_score(ytest, rd_pred)
print(f"Accuracy Score of Random Forest is : {rd_acc}")
print(f"Classification Report : \n{clf_rd}")
```

```python
#Support Vector Machine Model
from sklearn.svm import SVC
svm = SVC()
svm.fit(Xtrain, ytrain)
```

```python
svm_pred = svm.predict(Xtest)
```

```python
clf_svm = classification_report(ytest, svm_pred)
svm_acc = accuracy_score(ytest, svm_pred)
print(f"Accuracy Score of svm is : {svm_acc}")
print(f"Classification Report : \n{clf_svm}")
```
